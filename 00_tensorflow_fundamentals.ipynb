**Creating random tensors**

Random tensors are tensors of some arbitrary size which contain random numbers.

```python
# Create two random (but the same) tensors
random_1 = tf.random.Generator.from_seed(42) # set seed for reproducibility
random_1 = random_1.normal(shape=(3,2))
random_2 = tf.random.Generator.from_seed(42) # set seed for reproducibility
random_2 = random_2.normal(shape=(3,2))

# Are they equal
random_1, random_2, random_1 == random_2
```

**Getting information from tensors**

When dealing with tensors you probably want to be aware of the followning attributes:

*    Shape
*    Rank
*    Axis or Dimension
*    Size







```python
# create a rank 4 tensor(we want four dimension)
rank_4_tensor = tf.zeros(shape = (2,3,4,5))
rank_4_tensor
```

```python
rank_4_tensor[0]
```

```python
rank_4_tensor.shape, rank_4_tensor.ndim, tf.size(rank_4_tensor)
```

```python
# Get various attributes of our tensor
print("Data type of every element :",rank_4_tensor.dtype)
print("Number of dimensions (rank) :",rank_4_tensor.ndim)
print("Shape of tensor :",rank_4_tensor.shape)
print("Elements along the 0 axis :",rank_4_tensor.shape[0])
print("Elements along the last axis :",rank_4_tensor.shape[-1])
print("Total number of elements :",tf.size(rank_4_tensor))
print("Total number of elements :",tf.size(rank_4_tensor).numpy())
```

```python
# Create a random tensor with values between 0 and 100 of size 50
tensor = tf.constant(np.random.randint(0,100,size = 50))
tensor
```

```python
# Find the min
tf.reduce_min(tensor)
```

```python
# Find the max
tf.reduce_max(tensor)
```

```python
# find the mean
tf.reduce_mean(tensor)
```

```python
# find the sum
tf.reduce_sum(tensor)
```

```python
# find the variance of our tensor
# tf.reduce_variance(tensır) will not work
```

```python
# To find the varience of our tensor, we need access to tensorflow_probability
import tensorflow_probability as tfp
tfp.stats.variance(tensor)
```

```python
# Find the standart deviation
tf.math.reduce_std(tf.cast(tensor,dtype = tf.float32))
# do not forget this usage
```

```python
# another way of finding variance
tf.math.reduce_variance(tf.cast(tensor, dtype = tf.float32))
```

Find the positional maximum and minimum

```python
# Create a new tensor for finding positional minimum and maximum
tf.random.set_seed(42)
f = tf.random.uniform(shape = [50])
f
```

```python
# Find the positional maximum
tf.argmax(f)
```

```python
# Index on our largest value position
f[tf.argmax(f)]
```

```python
# Find the max
tf.reduce_max(f)
```

```python
# Check for equality
tf.reduce_max(f) == f[tf.argmax(f)]
```

```python
# Find the positional minimum
tf.argmin(f)
```

```python
# Index on our smallest value position
f[tf.argmin(f)]
```
